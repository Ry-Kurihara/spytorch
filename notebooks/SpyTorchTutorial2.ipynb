### モジュールインポート

```python
import os

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns

import torch
import torch.nn as nn
import torchvision
```

#### ネットワークパラメータの定義

```python
# The coarse network structure is dicated by the Fashion MNIST dataset. 
nb_inputs  = 28*28
nb_hidden  = 100
nb_outputs = 10

time_step = 1e-3
nb_steps  = 100

batch_size = 256
```

```python
dtype = torch.float

# Check whether a GPU is available
if torch.cuda.is_available():
    device = torch.device("device=cuda")     
else:
    device = torch.device("device=cpu")
```

#### FashionMnistデータセットのダウンロード

```python
# Here we load the Dataset
root = os.path.expanduser("~/data/datasets/torch/fashion-mnist")
train_dataset = torchvision.datasets.FashionMNIST(root, train=True, transform=None, target_transform=None, download=True)
test_dataset = torchvision.datasets.FashionMNIST(root, train=False, transform=None, target_transform=None, download=True)
```

```python
#Pytorchのアップデートで.train_data, .test_data => .data になり、 .train_labels, .test_labels => .target と記述するほうが適切になったぽい

x_train = np.array(train_dataset.data, dtype=np.float)#size[60000, 28, 28]
x_train = x_train.reshape(x_train.shape[0],-1)/255#reshape(6000, -1)によってサイズは[6000, 784]になる。各要素に対して255割る

x_test = np.array(test_dataset.data, dtype=np.float)#size[10000, 784]
x_test = x_test.reshape(x_test.shape[0],-1)/255#reshape(10000, -1)によってサイズは[10000, 784]になる。

y_train = np.array(train_dataset.targets, dtype=np.int) #サイズ[60000]
y_test  = np.array(test_dataset.targets, dtype=np.int) #サイズ[10000]
```

```python
# Here we plot one of the raw data points as an example
data_id = 1
plt.imshow(x_train[data_id].reshape(28,28), cmap=plt.cm.gray_r)
plt.axis("off")
```

Since we are working with spiking neural networks we ideally want to use a temporal code to make use of spike timing. To that end we will use a spike latency code to feed spikes to our network.

#### テンポラルコーディングによるデータエンコーディング

```python
def current2firing_time(x, tau=20, thr=0.2, tmax=1.0, epsilon=1e-7):
    idx = x<thr #x=x_trainのサイズ[60000][784]で、要素がTrue or False
    x = np.clip(x,thr+epsilon,1e9) #xの要素の最小値をthr+epsilon, 最大値を1e9に正規化する
    T = tau*np.log(x/(x-thr)) #サイズ[60000][784]で、xの要素を調整する
    T[idx] = tmax #Trueで発火。Falseは0にする。
    return T #サイズ[60000][784]

def sparse_data_generator(X, y, batch_size, nb_steps, nb_units, shuffle=True ):
    labels_ = np.array(y, dtype=np.int) #後ろアンダースコアはPythonのキーワードとの衝突を防ぐため。サイズ[60000]
    number_of_batches = len(X)//batch_size #//に注意！！結果が整数になる。小数点切り捨て。バッチ操作は、余りは使わないもの？？
    sample_index = np.arange(len(X))#array([1,2,3,4,5,6,.....])
  
    tau_eff = 20e-3/time_step#20ms/10ms=2
    firing_times = np.array(current2firing_time(X, tau=tau_eff, tmax=nb_steps), dtype=np.int)#[60000, 784]で帰ってくる
    unit_numbers = np.arange(nb_units)
  
    if shuffle:
        np.random.shuffle(sample_index)#samples_indexがarray([1,2,3,4,...])からarray([5,34,67,900,...])のようにランダムになる。
    
    total_batch_count = 0
    counter = 0
    while counter<number_of_batches:#すべてのバッチに対して処理を行う。トレーニングデータなら60000/256=234.375=234なので234回繰り返されるはず。
        batch_index = sample_index[batch_size*counter:batch_size*(counter+1)]#バッチサイズ分だけのデータを取り出すためのインデックス。サイズ[256]
    
        coo = [ [] for i in range(3) ]
        for bc,idx in enumerate(batch_index):#bcには[1,2,3....], idxに配列batch_indexの要素intが入る。
            c = firing_times[idx]<nb_steps #サイズ[784]で要素にTrue or Falseがある。発火時刻が時間窓を超えてしまった場合はFalse
            times, units = firing_times[idx][c], unit_numbers[c] #timesのサイズは[0~784]で発火時刻。 unitsのサイズは[0~784]でtimesのサイズと同じで、発火するユニットのインデックス
      
            batch = [bc for _ in range(len(times))]
            #extendはappendと違い、引数がリストで与えられ、実行オブジェクトの次元は維持したまま要素が加わる。
            coo[0].extend(batch) #[256 -> 512 -> 768]と要素が増えていく。最終的に60000まで
            coo[1].extend(times) #[0~768]データあるうちのnb_step以下だった要素。発火時刻がインデックスされている。範囲は[0*100 ~ 768*100]
            coo[2].extend(units) #サイズは同上。要素は発火するユニット番号
      
        i = torch.LongTensor(coo).to(device) #整数用のTensor。サイズは[[60000], [0*100 ~ 768*100], [左と同じ]]
        v = torch.FloatTensor(np.ones(len(coo[0]))).to(device) #サイズは[60000]
    
        #メモリ効率を考慮して疎行列オブジェクトに変換。torch.Size()で指定したサイズが大きいぶんには余ったところは0が要素となる？？このオブジェクトに対してto_dense()メソッドを用いることで通常の行列に戻す
        X_batch = torch.sparse.FloatTensor(i, v, torch.Size([batch_size, nb_steps, nb_units])).to(device) #インデックスの座標iと値vをペアで渡す。入力層ならtorch.Size([256, 100, 784])
        y_batch = torch.tensor(labels_[batch_index], device=device) #変数語のアンダーバーはPython既存変数との重複を防ぐ。サイズは、60000のうちの256のインデックスを取るので[256]。最後は96
    
        yield X_batch.to(device=device), y_batch.to(device=device) #returnにすると1ループ目で関数を抜ける。バッチを1つづつリターンで渡していく。xサイズ[256,100,784]の疎行列, yサイズ[256]のラベル
    
        counter += 1
```

### Setup of the spiking network model

### SNNモデルの構築

```python
tau_mem = 10e-3
tau_syn = 5e-3

alpha   = float(np.exp(-time_step/tau_syn))
beta    = float(np.exp(-time_step/tau_mem))
```

```python
weight_scale = 7*(1.0-beta) # this should give us some spikes to begin with

w1 = torch.empty((nb_inputs, nb_hidden),  device=device, dtype=dtype, requires_grad=True)
torch.nn.init.normal_(w1, mean=0.0, std=weight_scale/np.sqrt(nb_inputs))

w2 = torch.empty((nb_hidden, nb_outputs), device=device, dtype=dtype, requires_grad=True)
torch.nn.init.normal_(w2, mean=0.0, std=weight_scale/np.sqrt(nb_hidden))

print("init done")
```

```python
def plot_voltage_traces(mem, spk=None, dim=(3,5), spike_height=5):
  gs = GridSpec(*dim)
  if spk is not None:#発火しているとき
    dat = (mem+spike_height*spk).detach().cpu().numpy()#height*spk分だけ高くプロットし発火をわかりやすくする。
  else:
    dat = mem.detach().cpu().numpy()
  for i in range(np.prod(dim)):
    if i==0: a0=ax=plt.subplot(gs[i])
    else: ax=plt.subplot(gs[i], sharey=a0)
    ax.plot(dat[i])
    ax.axis("off")
```

We can now run this code and plot the output layer "membrane potentials" below. As desired these potentials do not have spikes riding on them.

## Training the network

### ネットワークの訓練

#### 伝搬を自分で定義

```python
class SuperSpike(torch.autograd.Function):
  scale = 100.0
  
  @staticmethod
  def forward(ctx, input):
    ctx.save_for_backward(input)
    out = torch.zeros_like(input)
    out[input > 0] = 1.0
    return out
  
  @staticmethod
  def backward(ctx, grad_output):
    input, = ctx.saved_tensors
    grad_input = grad_output.clone()
    grad = grad_input/(SuperSpike.scale*torch.abs(input)+1.0)**2 # **2で2乗を表す。
    return grad
  
spike_fn = SuperSpike.apply
```

#### 順伝搬関数を定義

```python
def run_snn(inputs):
    h1 = torch.einsum("abc,cd->abd", (inputs, w1))
    syn = torch.zeros((batch_size,nb_hidden), device=device, dtype=dtype)
    mem = torch.zeros((batch_size,nb_hidden), device=device, dtype=dtype)

    mem_rec = [mem]
    spk_rec = [mem]

    # Compute hidden layer activity
    for t in range(nb_steps):
        mthr = mem-1.0
        out = spike_fn(mthr)
        rst = torch.zeros_like(mem)
        c   = (mthr > 0)
        rst[c] = torch.ones_like(mem)[c]

        new_syn = alpha*syn +h1[:,t]
        new_mem = beta*mem +syn -rst

        mem = new_mem
        syn = new_syn

        mem_rec.append(mem)
        spk_rec.append(out)

    mem_rec = torch.stack(mem_rec,dim=1)
    spk_rec = torch.stack(spk_rec,dim=1)

    # Readout layer
    h2= torch.einsum("abc,cd->abd", (spk_rec, w2))
    flt = torch.zeros((batch_size,nb_outputs), device=device, dtype=dtype)
    out = torch.zeros((batch_size,nb_outputs), device=device, dtype=dtype)
    out_rec = [out]
    for t in range(nb_steps):
        new_flt = alpha*flt +h2[:,t]
        new_out = beta*out +flt

        flt = new_flt
        out = new_out

        out_rec.append(out)

    out_rec = torch.stack(out_rec,dim=1)
    other_recs = [mem_rec, spk_rec]
    return out_rec, other_recs
```

#### トレーニングとテストの関数定義

```python
def train(x_data, y_data, lr=2e-3, nb_epochs=10):
    params = [w1,w2]
    optimizer = torch.optim.Adam(params, lr=lr, betas=(0.9,0.999))

    log_softmax_fn = nn.LogSoftmax(dim=1)
    loss_fn = nn.NLLLoss()
    
    loss_hist = []
    for e in range(nb_epochs):
        local_loss = []
        for x_local, y_local in sparse_data_generator(x_data, y_data, batch_size, nb_steps, nb_inputs): #xサイズ[256,100,784]の疎行列, yサイズ[256]のラベルを受け取る。受け取る回数は60000//256=234.375=234なので234回。
            #.to_dense()で密行列に直す。サイズは[256][100][784]となる。：：：：：参照ー＞README.md
            output,_ = run_snn(x_local.to_dense())
            m,_= torch.max(output,1) # mに100time中で最大の膜電位を代入、_に100timのなかで最大の膜電位を持つインデックスを代入。サイズ[256,10]
            log_p_y = log_softmax_fn(m) #[256,10]サイズの最大膜電位に対してソフトマックス関数を適用。log(Softmax(x))関数を通して出てくる。サイズは同じ[256,10]。dimを変えてもサイズは同じで数値が変わって出てくる。
            loss_val = loss_fn(log_p_y, y_local) #スカラとして出てくる。例(0.7294）

            optimizer.zero_grad() #gradを初期化
            loss_val.backward()
            optimizer.step()
            local_loss.append(loss_val.item())
        mean_loss = np.mean(local_loss)
        print("Epoch %i: loss=%.5f"%(e+1,mean_loss))
        loss_hist.append(mean_loss)
        
    return loss_hist
        
        
def compute_classification_accuracy(x_data, y_data):
    """ Computes classification accuracy on supplied data in batches. """
    accs = []
    for x_local, y_local in sparse_data_generator(x_data, y_data, batch_size, nb_steps, nb_inputs, shuffle=False):
        output,_ = run_snn(x_local.to_dense())
        m,_= torch.max(output,1) # max over time
        _,am=torch.max(m,1)      # argmax over output units
        tmp = np.mean((y_local==am).detach().cpu().numpy()) # compare to labels
        accs.append(tmp)
    return np.mean(accs)
```

#### トレーニング

```python
loss_hist = train(x_train, y_train, lr=2e-4, nb_epochs=30)
```

```python
plt.figure(figsize=(3.3,2),dpi=150)
plt.plot(loss_hist)
plt.xlabel("Epoch")
plt.ylabel("Loss")
sns.despine()
```

#### テスト

```python
print("Training accuracy: %.3f"%(compute_classification_accuracy(x_train,y_train)))
print("Test accuracy: %.3f"%(compute_classification_accuracy(x_test,y_test)))
```

```python
def get_mini_batch(x_data, y_data, shuffle=False):
    for ret in sparse_data_generator(x_data, y_data, batch_size, nb_steps, nb_inputs, shuffle=shuffle):
        return ret 
```

```python
x_batch, y_batch = get_mini_batch(x_test, y_test)
output, other_recordings = run_snn(x_batch.to_dense())
mem_rec, spk_rec = other_recordings
```

In conclusion we see that already this simple spiking network solves the classification problem with ~85% accuracy and there is plenty of room left for tweaking. However, the hidden layer activities do not look very biological. Although the network displays population sparseness in that only a subset of neurons are active at any given time, the individual neurons' firing rates are pathologically high. This is not too surprising since we have not incentivised low activity levels in any way. We will create such an incentive to address this issue by activity regularization in one of the next tutorials.
